```python
import pandas as pd
import numpy as np
import datetime as dt
from sklearn.model_selection import train_test_split
```

```python
X=pd.read_csv('top30_2.csv')
```

```python
X.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
X.head()
```

```python
fdr_card=X[(X['Date']>='2010-11-01')&(X['Cardnum']==5142235211)]
fdr_card.head()
```

```python
fdr_merch=X[(X['Date']>='2010-11-01')&(X['Merchnum']=='4353000719908')]
fdr_merch.head()
```

```python
oot_x1=fdr_card.drop(['Cardnum','Date','Merchnum','Fraud'],axis=1)
oot_x2=fdr_merch.drop(['Cardnum','Date','Merchnum','Fraud'],axis=1)
```

```python
y=X['Fraud']
```

```python
X=X.drop(['Cardnum','Date','Merchnum','Fraud'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=150,max_depth=20,n_jobs=5)
model.fit(X_train,y_train.values.ravel())
```

```python
fdr_card['Fraud Score']=model.predict_proba(oot_x1)[:,1]
fdr_merch['Fraud Score']=model.predict_proba(oot_x2)[:,1]
```
